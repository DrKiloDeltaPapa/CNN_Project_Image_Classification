```python
import os

# Define the source folder
source_folder = "/home/locutus/Downloads/CNN_Project_Group_17/dataset/test/Animals"  # Path to the "Animals" folder

# List all files in the folder
images = os.listdir(source_folder)

# Sort the files in case the order matters (e.g., alphabetical)
images.sort()

# Rename the images with the "animal_" prefix followed by the image number
for i, img in enumerate(images):
    # Ensure it's an image file (optional check, in case there are non-image files)
    if img.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Get the file extension (e.g., .jpg, .png)
        extension = img.split('.')[-1]
        # Construct the new name (e.g., animal_1.jpg, animal_2.jpg)
        new_name = f"animal_{i + 1}.{extension}"

        # Rename the image
        os.rename(os.path.join(source_folder, img), os.path.join(source_folder, new_name))

        print(f"Renamed: {img} -> {new_name}")

print("Renaming complete!")
```
